## Data Understanding

The data includes over 1000 rows from the 21/22 and 22/23 season, where each row is a single match played.

```python
# Import standard packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# read in the data
df = pd.read_csv("./data/matches.csv", index_col=0)
```

```python
# create a function that performs simple EDA
def explore_data(dataframe):
    # Summary statistics
    summary_stats = dataframe.describe()
    
    # Data type information
    data_types = dataframe.dtypes
    
    # Missing value analysis
    missing_values = dataframe.isnull().sum()
    missing_percentage = (missing_values / len(dataframe)) * 100
    missing_data = pd.DataFrame({'Missing Count': missing_values, 'Missing Percentage': missing_percentage})
    
    # Value distribution for each column
    team_vc = dataframe.team.value_counts()
    
    # Display results
    print("Summary Statistics:")
    print(summary_stats)
    print("\nData Types:")
    print(data_types)
    print("\nMissing Values:")
    print(missing_data)
    print("\nGames per team:")
    print(team_vc)
```

```python
explore_data(df)
```

```python
# create a function that generates histograms for nominal values 
# to check distributions
def plot_histograms(dataframe):
    # Select numeric columns
    numeric_cols = dataframe.select_dtypes(include=['int64', 'float64']).columns
    
    # Plot histograms
    for column in numeric_cols:
        plt.figure(figsize=(8, 6))
        sns.histplot(dataframe[column], kde=True, color='skyblue')
        plt.title(f"Histogram of '{column}'")
        plt.xlabel(column)
        plt.ylabel("Frequency")
        plt.show()
```

```python
plot_histograms(df)
```
